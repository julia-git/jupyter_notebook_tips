## Tip #5 - Built-in Magic Commands

Here are some other useful magic commands:

```python
%pwd #print the current working directory
```

```python
%cd #change working directory
```

```python
%ls #show contents in the current directory
```

```python
%load [insert Python filename here] #load code into the Jupyter notebook
```

```python
%store [insert variable here] #this lets you pass variables between Jupyter Notebooks
```

```python
%who #use %who to list all variables
```

```python
a = 5 
b = 1
%who int
```

For the full list of magic commands:

```python
%lsmagic #show all magic commands
```

## Tip #8 - Alarm when code finishes

### For Windows:

```python
import winsound
duration = 1000  #milliseconds
freq = 440  #Hz
winsound.Beep(freq, duration)
```

### For Mac:

```python
import os
os.system('say "your program has finished"')
```

You might need to install the following:

sudo apt install speech-dispatcher

Source: https://stackoverflow.com/questions/16573051/sound-alarm-when-code-finishes/16573339#16573339
